```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a2c = pd.read_csv('run-A2C_1-tag-rollout_ep_rew_mean.csv')
dqn = pd.read_csv('run-DQN_1-tag-rollout_ep_rew_mean.csv')
ppo = pd.read_csv('run-PPO_1-tag-rollout_ep_rew_mean.csv')
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(a2c['Step'], a2c['Value'], label='A2C')
plt.plot(dqn['Step'], dqn['Value'], label='DQN')
plt.plot(ppo['Step'], ppo['Value'], label='PPO')
plt.xlabel('Training Episodes')
plt.ylabel('Reward')
plt.title('Reward vs Episodes for Deep RL Agents in Block 3 of Foraging Env')
plt.legend()
plt.show()
```
